```python
label = {'pop': 0, 'rnb': 1}
data = []
file_mapping=[]
for filename in os.listdir('/content/drive/My Drive/dataset/IR/POP'):
    songData = []
    songname = f'/content/drive/My Drive/dataset/IR/POP/{filename}'
    y, sr = librosa.load(songname, mono=True, duration=30)
    
    tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
    songData.append(tempo)
    songData.append(mean(beats))

    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    songData.append(mean(chroma_stft.flatten()))

    rmse=librosa.feature.rms(y=y)[0]
    songData.append(mean(rmse.flatten()))

    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    songData.append(mean(spec_cent.flatten()))

    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    songData.append(mean(spec_bw.flatten()))

    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    songData.append(mean(rolloff.flatten()))

    zcr = librosa.feature.zero_crossing_rate(y)
    songData.append(mean(zcr.flatten()))

    mfcc = librosa.feature.mfcc(y=y, sr=sr)
    for i in mfcc:
        songData.append(np.mean(i))
    
    songData.append(label['pop'])

    data.append(songData)
    
```

```python
folders = {
           'POP_20s':'/content/drive/My Drive/dataset/IR/IR_data/POP_20s',
           'RnB_20s':'/content/drive/My Drive/dataset/IR/IR_data/RnB_20s',
           'folk_20s':'/content/drive/My Drive/dataset/IR/IR_data/folk_20s',
           'hiphop_20s':'/content/drive/My Drive/dataset/IR/IR_data/hiphop_20s',
           'rock_20s':'/content/drive/My Drive/dataset/IR/IR_data/rock_20s'
           }
label = {'POP_20s': 0, 'RnB_20s': 1, 'folk_20s': 2, 'hiphop_20s': 3, 'rock_20s': 4}
data_20s = []
```

```python
itr=0
i1=0
for name, path in folders.items():
  for filename in os.listdir(path):
    songData = []
    songname = f'{path}/{filename}'
    print(songname)
    if(i1>=len(data_20s)):
      numbers = ''
      for word in filename:
        if word.isdigit():
          numbers=numbers+word
      numbers=int(int(numbers)/10)
      y, sr = librosa.load(songname, mono=True)
      try:
        tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
        songData.append(tempo)
        songData.append(mean(beats))

        chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
        songData.append(mean(chroma_stft.flatten()))

        rmse=librosa.feature.rms(y=y)[0]
        songData.append(mean(rmse.flatten()))

        spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
        songData.append(mean(spec_cent.flatten()))

        spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        songData.append(mean(spec_bw.flatten()))

        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        songData.append(mean(rolloff.flatten()))

        zcr = librosa.feature.zero_crossing_rate(y)
        songData.append(mean(zcr.flatten()))

        mfcc = librosa.feature.mfcc(y=y, sr=sr)
        for i in mfcc:
            songData.append(np.mean(i))

        songData.append(label[name])
        if itr==0:
          songData.append(filemap_POP[numbers])
        if(itr==1):
          songData.append(filemap_RnB[numbers])
        if(itr==2):
          songData.append(filemap_folk[numbers])
        if(itr==3):
          songData.append(filemap_hiphop[numbers])
        if(itr==4):
          songData.append(filemap_rock[numbers])
        print(numbers)
        data_20s.append(songData)
      except:
        print("None")
    i1+=1
  print(itr)
  itr+=1
```
