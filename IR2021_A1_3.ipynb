```
  #Function for preprocessing
  word_list = []
  unique_words_total = []

  def pre_processing(file_path, temp):
    tokens=[]
    
    if(temp==0):                                         ## parameter is a file
      rfile = open(file_path,'r',errors = 'ignore')
      read_file = rfile.read()
    else:                                               ## paramter is a query
      read_file=file_path

  # Tokenize files into sentence
    sent_tokens = sent_tokenize(read_file)   
    strings = ''  

  #Tokenize sentences into words
    for w in sent_tokens:
      word = w.split(' ')                    
      strings = ' '.join(word)   

      #Convert to lowercase
      strings = strings.lower()    

      #Expand contractions    
      s=contractions.fix(strings)    #NOT REMOVING appostopic s in Mark's

      #Remove punctuations and extra spaces
      clean_string =  re.sub(r'[\!@#$%^&\*()\_+={}\:\;<>\?/\|\,\-/"\']*',"",s)
      clean_string = re.sub('\n', ' ', clean_string)
      clean_string = re.sub('\s+', ' ', clean_string)
      clean_string = re.sub('^\s+', '', clean_string)
      clean_string=re.sub("[^a-zA-Z\s\n]", " ",clean_string)  
      clean_string = re.sub(r"(.)\1{2}"," ",clean_string)  #for elongated words

      #Remove Stop words
      stop_words = set(stopwords.words('english'))
      clean_word = clean_string.split()
      for x in clean_word:
        if x not in stop_words and len(x)>1:
          if x not in unique_words_total:
            unique_words_total.append(x)
          tokens.append(x)
    return tokens
```

```
doc = 1
tokens_in_file={}
file_words={}
for file_path in files_list:
  doc_id = ntpath.basename(file_path)                        
  tokens=pre_processing(file_path,0)
  tokens_in_file[doc]=tokens
  file_words[doc]=doc_id
  doc=doc+1
  

## lemmatisation of unique words obtained
lemmatizer = WordNetLemmatizer() 
for i in range(len(unique_words_total)):
  unique_words_total[i] = lemmatizer.lemmatize(unique_words_total[i])

### final unique words after lematization
unique_words_final=[]

for words in unique_words_total:
  if(words not in unique_words_final):
    unique_words_final.append(words)
print(len(unique_words_final))
print(unique_words_final)
```

```
class Node:
  def __init__(self, docId, freq=None):
    self.freq=freq
    self.doc=docId
    self.nextval=None
class LinkedList:
  def __init__(self, head=None):
    self.head=head
```

```
def merge(list1, list2, operator,comparisons):
  new_list=LinkedList()
  new_list.head=Node(0, Node)                                                   # Initially linked list is 0-> null
  temp=new_list.head

  if(operator=='and'):
    while list1 is not None and list2 is not None:               # AND Opertion
      comparisons = comparisons + 1
      if(list1.doc > list2.doc):
        list2=list2.nextval

      elif list1.doc<list2.doc:
        list1=list1.nextval
      else:
        temp.nextval=Node(list1.doc, Node)
        list1=list1.nextval
        list2=list2.nextval
        temp=temp.nextval
    return new_list,comparisons

  if(operator=='or'):
    while list1 is not None and list2 is not None:              # OR Operation
      comparisons = comparisons + 1
      if(list1.doc > list2.doc):
        temp.nextval=Node(list2.doc, Node)
        temp=temp.nextval
        list2=list2.nextval
      elif list1.doc<list2.doc:
        temp.nextval=Node(list1.doc, Node)
        temp=temp.nextval
        list1=list1.nextval
      else:
        temp.nextval=Node(list1.doc, Node)
        list1=list1.nextval
        list2=list2.nextval
        temp=temp.nextval
        #print(temp.doc)
    if(list1 is None):
      temp.nextval=list2
    else:
      temp.nextval=list1
    return new_list,comparisons

                
  if(operator=='and-not'):                                          #AND NOT Operation
    ##Finding A-B
    while list1 is not None and list2 is not None: 
      comparisons = comparisons + 1                             
      if (list1.doc > list2.doc):
        list2=list2.nextval

      elif list1.doc<list2.doc:
        temp.nextval=Node(list1.doc, Node)
        list1=list1.nextval
        temp=temp.nextval
        
      else:
        list1=list1.nextval
        list2=list2.nextval
    if list2 is None:
      temp.nextval=list1
    return new_list,comparisons

  if(operator=="or-not"):                                                           # OR NOT Operation [A or not B= U-(B-A)]
    temp_list,comparisons=merge(list2, list1, 'and-not',comparisons)             # temp_list = B-A
    l=LinkedList()
    l.head=Node(0, Node)                             # l = U
    temps=l.head
    for i in range(1, 468):
      temps.nextval=Node(i, Node)
      temps=temps.nextval
    return merge(l.head.nextval, temp_list.head.nextval, 'and-not',comparisons)  #l-temp_list


```
