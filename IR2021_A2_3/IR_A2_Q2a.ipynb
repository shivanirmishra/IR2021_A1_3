Preprocessing

```python
#Function for preprocessing
def pre_processing(file_path, temp):
  tokens=[]
  if(temp==0):                                         ## parameter is a file
    rfile = open(file_path,'r',errors = 'ignore')
    read_file = rfile.read().replace('\n', ' ')
  else:                                               ## paramter is a query
    read_file=file_path                     

  #convert to lowercase
  read_file = read_file.lower() 

  #Tokenisation
  sent_tokens = sent_tokenize(read_file)   
  strings = ''  
  
  #Tokenize sentences into words
  for w in sent_tokens:
    word = w.split(' ')
    strings = ' '.join(word)                       

    #Expand contractions    
    s=contractions.fix(strings)    #NOT REMOVING appostopic s in Mark's

    #Remove punctuations and extra spaces
    clean_string = re.sub(r'[\!@#$%^&\*()\_+={}\:\;<>\?/\|\,\-/"\']*',"",s)
    clean_string = re.sub('\n', ' ', clean_string)
    clean_string = re.sub('\s+', ' ', clean_string)
    clean_string = re.sub('^\s+', '', clean_string)
    clean_string = re.sub("([\-]{1,2})", " ",clean_string)
    clean_string = re.sub('([>|\-=+~]+)', ' ', clean_string)
    clean_string=re.sub("([\.\!]{2})", " ",clean_string)
    clean_string=re.sub("[^a-zA-Z\s\n]", " ",clean_string)  
    clean_string = re.sub(r"(.)\1{2}","",clean_string)
    
    #Remove stopwords from tokens
    clean_word = clean_string.split(' ')
    stop_words = set(stopwords.words('english'))
    for x in clean_word:
      if x not in stop_words and len(x)>1:
        tokens.append(x)
  return tokens
```

## **Q2a Jaccard Coefficient**

```python
query=input("Input the query")
```

```python
def doc_read(file_path):
  with open(file_path, 'r', errors='ignore') as file:
    doc1 = file.read().replace('\n', ' ')
  return doc1
```

```python
documents = []
file_name = {}
doc = 1
for file_path in files_list:
  doc_id = ntpath.basename(file_path)
  # if doc_id.endswith('.txt'):
  file_name[doc]=doc_id  
  # print(doc_id)
  doc = doc + 1
  documents.append(doc_read(file_path))
```

```python
tokens = []
sets = []
set_union = []
set_intersection = []
jac = []

for i in documents:
  # tokens.append([t.lower() for t in i.split(" ")])
  tokens.append(pre_processing(i, 1))
# tokens_query = [t.lower() for t in query.split(" ")]
tokens_query = pre_processing(query, 1)

for i in tokens:
  sets.append(set(i))
set_query = set(tokens_query)

for i in sets:
  set_union.append(i.union(set_query))
  set_intersection.append(i.intersection(set_query))

for (i,j) in zip(set_union, set_intersection):
  jac.append(len(j) / len(i))
```

```python
print(jac)
```

```python
dict_jacc = {}
temp=1
for i in jac:
  dict_jacc[temp] = i
  temp = temp+1
print(dict_jacc)

sorted_jacc = list(reversed(sorted(dict_jacc.items(), key = lambda kv:(kv[1], kv[0]))))
print(sorted_jacc)
```
